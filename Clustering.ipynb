```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

data = pd.read_csv('StateWiseRainfall.csv')
data
data['State']
data['Rainfall']


```

```python

plt.scatter(data['State'],data['Rainfall'])

plt.show()
```

```python
x = data.iloc[:,0:2]
print(x)
```

```python
kmeans = KMeans(5)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = data.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['State'],data_with_clusters['Rainfall'],c=data_with_clusters['Clusters'],cmap='rainbow')
```

```python

import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
data = pd.read_csv('State.csv')
```

```python
plt.scatter(data['Rainfall'],data['Longitude'])

plt.show()
```

```python
x = data.iloc[:,1:3]
```

```python
kmeans = KMeans(5)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
data_with_clusters = data.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['Rainfall'],data_with_clusters['Longitude'],c=data_with_clusters['Clusters'],cmap='rainbow')
```

```python
wcss=[]
for i in range(1,7):
  kmeans = KMeans(i)
  kmeans.fit(x)
  wcss_iter = kmeans.inertia_
  wcss.append(wcss_iter)

number_clusters = range(1,7)
plt.plot(number_clusters,wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```
